```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
# Define the equation as a symbolic expression
x = sp.symbols('x')
equation = sp.exp(x) - 3*x
```

```python
def eq_function(x_value):
    result=equation.subs(x, x_value)
    return result
```

```python
def newton_raphson(x0, tolerance, max_iterations):
    x_values = [x0]
    iterations = [0]

    for i in range(max_iterations):
        x1 = x0 - eq_function(x0) / sp.diff(equation, x).subs(x, x0)
        print(f"Result Of x in each iteration : {x1}")
        x_values.append(x1)
        iterations.append(i + 1)
        if abs(x1 - x0) < tolerance:
            break
        x0 = x1

    return x_values, iterations
```

```python
# Prompt the user for input
x0 = float(input("Enter the initial guess 'x0': "))
tolerance = 1e-6
max_iterations = 100
x_values, iterations = newton_raphson(x0, tolerance, max_iterations)
```

```python
# Plot the convergence
plt.plot(iterations, x_values, marker='o', linestyle='-')
plt.xlabel("Iterations")
plt.ylabel("x")
plt.title("Newton-Raphson Method Convergence")
plt.grid(True)
plt.show()
```
